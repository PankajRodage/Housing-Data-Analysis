```python
# 4.create a histogram of sale price also showing mean, median and mode
plt.figure(figsize=(6,4))
sns.histplot(data=df,x='SalePrice',color='pink',edgecolor='yellow')
plt.vlines(x=df['SalePrice'].mean(),ymin=0,ymax=350,color='r',label='Mean')
plt.vlines(x=df['SalePrice'].median(),ymin=0,ymax=350,color='b',label='Median')
plt.vlines(x=df['SalePrice'].mode(),ymin=0,ymax=350,color='k',label='Mode')
plt.show()
```

```python
# 5.box plot for sale price
plt.figure(figsize=(6,4))
plt.boxplot(df['SalePrice'])
plt.show()
```

```python
# 6.create a histogram lot area also showing mean, median and mode
plt.figure(figsize=(6,4))
sns.histplot(data=df,x='Lot Area',color='orange',edgecolor='yellow')
plt.vlines(x=df['Lot Area'].mean(),ymin=0,ymax=250,color='r',label='Mean')
plt.vlines(x=df['Lot Area'].median(),ymin=0,ymax=250,color='k',label='Median')
plt.vlines(x=df['Lot Area'].mode(),ymin=0,ymax=250,color='g',label='Mode')
plt.show()
```

```python
# 6. box plot of lot area
plt.figure(figsize=(6,4))
plt.boxplot(df['Lot Area'])
plt.show()
```

```python
# 7.scatter plot of lot area and sale price
sns.scatterplot(x=df['Lot Area'],y=df['SalePrice'],data=df,palette='viridis',sizes=(5,100))
plt.show()
```

```python
# 8. Create a new feature (column) called Price per Area in the data frame to contain ratio of "SalePrice" and "Lot Area" features
df['Price per Area ']= df['SalePrice'] / df['Lot Area']
```

```python
# 9. Check if the new column is created using head()
df.head()
```

```python
# 10. Create a histogram of the "Price per Area" feature and also showing the mean and the median
df['Price per Area'] = df['SalePrice'] / df['Lot Area']
plt.figure(figsize=(6,4))
sns.histplot(data=df,x='Price per Area',color='orange',edgecolor='yellow')
plt.vlines(x=df['Price per Area'].mean(),ymin=0,ymax=350,color='b',label='Mean')
plt.vlines(x=df['Price per Area'].median(),ymin=0,ymax=350,color='k',label='Median')
plt.vlines(x=df['Price per Area'].mode(),ymin=0,ymax=350,color='g',label='Mode')
plt.show()
```

```python
# 10. box plot of the "Price per Area"
df['Price per Area'] = df['SalePrice'] / df['Lot Area']
plt.figure(figsize=(6,4))
plt.boxplot(df['Price per Area'])
plt.show()
```

```python
# 11. Create a scatter plot between lot area and price per area
df['Price per Area'] = df['SalePrice'] / df['Lot Area']
sns.scatterplot(x=df['Lot Area'],y=df['Price per Area'],data=df,palette='viridis',sizes=(5,100))
plt.show()
```
